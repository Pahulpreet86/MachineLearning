```python
#simple regression model
#import library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
df = pd.read_csv("House Sales in King County, USA.csv") #Reading the dataset (dataframe) 
df.head()#shows the top 5 rows
```

```python
df.columns #data dictionary (name of all the columns)
#dependent variable
```

```python
df.apply(lambda x: sum(x.isnull()),axis=0) #check missing value isnull() return 1 for missing value
```

```python
df.describe() #summarization of data frame
```

```python
df.corr() #find correlation between different factors and price 
#observed -> max correlation of price with sqft_living
```

```python
Y_dataset=df.iloc[:, 2].values #dependent variable #price column
```

```python
X_dataset=df.iloc[ : ,5 : 6].values #independent variable #sqft_living space column 
```

```python
#Split dataset for training and testing
X_train, X_test, y_train, y_test = train_test_split(X_dataset,Y_dataset,test_size=0.1 , random_state=10) #test size=0.1 -> 90% train and 10% test
```

```python
simple_linear_regression=LinearRegression() #define 
simple_linear_regression.fit(X_train,y_train) #fit model 
```

```python
#check accuracy of model
score=simple_linear_regression.score(X_test, y_test)
print "Accuracy= ",(score*100),"%"
```

```python
#data visualization 
#Visualizing the Test Results 
plt.scatter(X_test, y_test, color= 'red',label='Actual')
plt.title ("Result Visualization ")
plt.scatter(X_test,simple_linear_regression.predict(X_test),color= 'blue',label='Predicted')
plt.xlabel("Space")
plt.ylabel("Price")
plt.legend(loc="bottom right")
plt.show()
```
