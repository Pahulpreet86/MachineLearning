```python
#logistic regression
#predict chocolate or not
#simple regression model
#import library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
df = pd.read_csv("The Ultimate Halloween Candy Power Ranking.csv") #Reading the dataset (dataframe) 
df.head()#shows the top 5 rows
```

```python
df.columns
```

```python
df.apply(lambda x: sum(x.isnull()),axis=0) #check missing value isnull() return 1 for missing value
```

```python
X_dataset=df.iloc[:, 2:].values#dependent variable 
Y_dataset=df.iloc[ : ,  1:2 ].values #all independent variable  
```

```python
#modeling using backward elimination
X_dataset_opt=np.append(arr= np.ones((len(df),1)).astype(int) ,values=X_dataset, axis =1)#adding a column with value all 1 as statsmodel doesnt take constant b0 in account
```

```python
order_least_square=sm.OLS(endog= Y_dataset, exog= X_dataset_opt).fit() #output(dependent variable) and input (independent) [fit model]
order_least_square.summary()#lower the p value ,more the variable significant SL=0.05 
```

```python
X_dataset_opt= X_dataset_opt[:, [0,1,2,3,7,10,11]]
order_least_square=sm.OLS(endog= Y_dataset, exog= X_dataset_opt).fit() #output(dependent variable) and input (independent) [fit model]
order_least_square.summary()#lower the p value ,more the variable significant SL=0.05 
```

```python
X_dataset_opt=X_dataset_opt[:, [1,2,3,7,10,11]]
```

```python
#Split dataset for training and testing
X_train, X_test, y_train, y_test = train_test_split(X_dataset_opt,Y_dataset,test_size=0.1 , random_state=0) #test size=0.1 -> 90% train and 10% test
```

```python
#define model
classifier=LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
```

```python
#check accuracy of model
score=classifier.score(X_test, y_test)
print "Accuracy= ",(score*100),"%"
```
